## Deliverable 1: Scrape Titles and Preview Text from Mars News

````python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup

````

````python
browser = Browser('chrome')
````

### Step 1: Visit the Website

1. Use automated browsing to visit the [Mars news site](https://static.bc-edx.com/data/web/mars_news/index.html). Inspect the page to identify which elements to scrape.

      > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools.

````python
# Visit the Mars news site
url = 'https://static.bc-edx.com/data/web/mars_news/index.html'
browser.visit(url)
````

### Step 2: Scrape the Website

Create a Beautiful Soup object and use it to extract text elements from the website.

````python
# Create a Beautiful Soup object
html = browser.html  # Get the HTML content of the page visited with the browser
soup_obj = soup(html, 'html.parser')  # Create a BeautifulSoup object

# Now you can use the soup_obj to extract text elements from the website.
````

````python
# Extract all text elements
all_text_elements = soup_obj.get_text()

# This will give you all the text on the webpage, including text from tags, scripts, and other elements.

print(all_text_elements)
````

````python
print(soup_obj)
````

### Step 3: Store the Results

Extract the titles and preview text of the news articles that you scraped. Store the scraping results in Python data structures as follows:

* Store each title-and-preview pair in a Python dictionary. And, give each dictionary two keys: `title` and `preview`. An example is the following:

  ```python
  {'title': "NASA's MAVEN Observes Martian Light Show Caused by Major Solar Storm", 
   'preview': "For the first time in its eight years orbiting Mars, NASA’s MAVEN mission witnessed two different types of ultraviolet aurorae simultaneously, the result of solar storms that began on Aug. 27."
  }
  ```

* Store all the dictionaries in a Python list.

* Print the list in your notebook.

````python
#  Create an empty list to store the dictionaries
results_list = []

# Find all elements with class "content_title" (assuming this corresponds to the titles)
titles = soup_obj.find_all(class_="content_title")

# Find all elements with class "article_teaser_body" (assuming this corresponds to the preview text)
previews = soup_obj.find_all(class_="article_teaser_body")
````

````python
# Loop through the titles and previews
for title, preview in zip(titles, previews):
    # Extract the text content from the title and preview elements
    extracted_title = title.get_text()
    extracted_preview = preview.get_text()

    # Create a dictionary for each news article
    news_item = {
        'title': extracted_title,
        'preview': extracted_preview
    }

    # Append the news item dictionary to the results list
    results_list.append(news_item)

# After the loop, you will have a list of dictionaries with titles and previews

````

````python
# Print the list to confirm success

for index, item in enumerate(results_list, start=1):
    print(f"Article {index}:")
    print(f"Title: {item['title']}")
    print(f"Preview: {item['preview']}\n")
````
